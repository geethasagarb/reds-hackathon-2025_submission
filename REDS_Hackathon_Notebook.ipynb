```python
from IPython.display import display

# Convert 'game_date' to datetime format
df_savan['game_date'] = pd.to_datetime(df_savan['game_date'])

# Extract the season (year)
df_savan['Year'] = df_savan['game_date'].dt.year

# Aggregate data per player per season
player_summary = df_savan.groupby(['batter', 'Year']).agg({
    'game_date': 'count',  # Count games played as a proxy for PA
    'events_walk': 'sum',
    'events_strikeout': 'sum',
    'events_home_run': 'sum',
    'description_hit_into_play': 'sum',

}).reset_index()

# Rename column for clarity
player_summary.rename(columns={'game_date': 'PA_Count'}, inplace=True)


display(player_summary)
```

```python
from IPython.display import display
# Aggregate BF (Batters Faced) for each pitcher per season
pitcher_summary = df_savan.groupby(['pitcher', 'Year']).agg({
    'game_date': 'count',  # Count plate appearances against the pitcher (proxy for BF)'events_walk': 'sum',  # Walks count as BF
    'events_strikeout': 'sum',  # Strikeouts count as BF
    'events_home_run': 'sum',  # Home runs count as BF
    'description_hit_into_play': 'sum',  # Balls in play count as BF

}).reset_index()

# Rename 'game_date' to 'BF_Count' to represent Batters Faced
pitcher_summary.rename(columns={'game_date': 'BF_Count'}, inplace=True)

display(pitcher_summary)
```

```python
batter_summary_renamed = player_summary.rename(columns={'batter': 'player_id', 'PA_Count': 'PA'})
pitcher_summary_renamed = pitcher_summary.rename(columns={'pitcher': 'player_id', 'BF_Count': 'BF'})

# Full Outer Join on 'player_id' and 'Year'
player_pitcher_summary = pd.merge(
    batter_summary_renamed, pitcher_summary_renamed,
    on=['player_id', 'Year'], how='outer'
)

# Fill missing values with 0 (some players are only batters, some only pitchers)
player_pitcher_summary.fillna(0, inplace=True)

# Display the merged dataset
from IPython.display import display
display(player_pitcher_summary[["player_id","Year","PA","BF"]])
```

```python
# Aggregate data per player per season using explicit PA count
from IPython.display import display, HTML
player_summary = df_savan.groupby(['batter', 'Year']).agg({
    'events_walk': 'sum',
    'events_strikeout': 'sum',
    'events_home_run': 'sum',
    'description_hit_into_play': 'sum',
}).reset_index()

# Sum total plate appearances explicitly
player_summary['PA_Count'] = player_summary[
    ['events_walk', 'events_strikeout', 'events_home_run', 'description_hit_into_play']
].sum(axis=1)

# Rename columns for clarity
player_summary.rename(columns={'batter': 'player_id', 'PA_Count': 'PA'}, inplace=True)
display(player_summary)
```

```python
# Aggregate BF (Batters Faced) per pitcher per season using explicit BF count
from IPython.display import display, HTML
pitcher_summary = df_savan.groupby(['pitcher', 'Year']).agg({
    'events_walk': 'sum',
    'events_strikeout': 'sum',
    'events_home_run': 'sum',
    'description_hit_into_play': 'sum',
}).reset_index()

# Sum total batters faced explicitly
pitcher_summary['BF_Count'] = pitcher_summary[
    ['events_walk', 'events_strikeout', 'events_home_run', 'description_hit_into_play']
].sum(axis=1)

# Rename columns for clarity
pitcher_summary.rename(columns={'pitcher': 'player_id', 'BF_Count': 'BF'}, inplace=True)
display(pitcher_summary)
```

```python
# Drop columns with more than 90% missing values
high_missing_cols = ['on_3b', 'on_2b', 'on_1b', 'sv_id']
df_savan = df_savan.drop(columns=high_missing_cols)
```

```python
# Fill categorical missing values with "unknown"
df_savan[['pitch_type', 'bb_type', 'hit_location']] = df_savan[['pitch_type', 'bb_type', 'hit_location']].fillna('unknown')

# Fill numerical missing values with median
numerical_fill_cols = ['release_speed', 'release_pos_x', 'release_pos_z', 'zone']
for col in numerical_fill_cols:
    df_savan[col] = df_savan[col].fillna(df_savan[col].median())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_df = df_savan.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, linewidths=0.5)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.title("Feature Correlation Heatmap", fontsize=14)
plt.show()
```

```python
# Select only correlations above 0.9 or below -0.9
high_corr_threshold = 0.9
high_corr = corr_matrix[abs(corr_matrix) > high_corr_threshold]

# Display highly correlated feature pairs
import pandas as pd
high_corr_pairs = high_corr.unstack().reset_index()
high_corr_pairs.columns = ['Feature_1', 'Feature_2', 'Correlation']

# Remove self-correlations (where Feature_1 == Feature_2)
high_corr_pairs = high_corr_pairs[high_corr_pairs['Feature_1'] != high_corr_pairs['Feature_2']]

# Drop duplicate pairs
high_corr_pairs = high_corr_pairs.sort_values(by='Correlation', ascending=False).drop_duplicates(subset=['Correlation'])

# Display correlated feature pairs
from IPython.display import display, HTML
display(HTML(high_corr_pairs.to_html()))
```

```python
from sklearn.preprocessing import LabelEncoder

# List of ordinal categorical features (values that have a ranking/order)
ordinal_features = ['stand', 'p_throws', 'game_type', 'inning_topbot', 'role_key']

# Apply Label Encoding to ordinal features
label_encoders = {}
for col in ordinal_features:
    label_encoders[col] = LabelEncoder()
    df_savan[col] = label_encoders[col].fit_transform(df_savan[col])
```

```python
# List of nominal categorical features (unordered)
nominal_features = ['pitch_type', 'batter', 'pitcher', 'home_team', 'away_team', 'type',
                    'hit_location', 'bb_type', 'fielder_2', 'pitch_name',
                    'if_fielding_alignment', 'of_fielding_alignment']

# Apply One-Hot Encoding
df_savan = pd.get_dummies(df_savan, columns=nominal_features, drop_first=True)
```

```python
# Identify DateTime columns
datetime_columns = df_savan.select_dtypes(include=['datetime64']).columns.tolist()
print("DateTime Columns:", datetime_columns)
```

```python
# Convert DateTime column into separate numerical features
df_savan['game_year'] = df_savan['game_date'].dt.year
df_savan['game_month'] = df_savan['game_date'].dt.month
df_savan['game_day'] = df_savan['game_date'].dt.day
df_savan['game_weekday'] = df_savan['game_date'].dt.weekday

# Drop the original DateTime column
df_savan = df_savan.drop(columns=['game_date'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE

# Select numeric features again (excluding target variables)
feature_columns = [col for col in df_savan.columns if col not in ['PA', 'BF']]

# Define input features and target variables
X = df_savan[feature_columns]
y_PA = df_savan['PA']
y_BF = df_savan['BF']

# Initialize model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Apply RFE for PA prediction
rfe_PA = RFE(estimator=model, n_features_to_select=20)
rfe_PA.fit(X, y_PA)
selected_features_PA = X.columns[rfe_PA.support_]

# Apply RFE for BF prediction
rfe_BF = RFE(estimator=model, n_features_to_select=20)
rfe_BF.fit(X, y_BF)
selected_features_BF = X.columns[rfe_BF.support_]

# Display selected features
print("Selected Features for PA Prediction:", list(selected_features_PA))
print("Selected Features for BF Prediction:", list(selected_features_BF))
```
